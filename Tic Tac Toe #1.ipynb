```python
#Display the table 
from IPython.display import clear_output
#to clear old outputs from the notebook 

sampletable = [' ']*10
#first one stays empty 

def display (table):
    clear_output()

    print(f"""
    {table[7]}  |  {table[8]}  |  {table[9]}\n
    {table[4]}  |  {table[5]}  |  {table[6]}\n
    {table[1]}  |  {table[2]}  |  {table[3]}\n
    """)
     
```

```python
#Player 1 and player 2 marker 

def ass_marker ():
    
    mark1 = ''
    player1 = ''
    player2 = ''
    
    while mark1.isdigit() or mark1!='X' and mark1!='O':
        
        mark1 = (input("P1: Would you like to be X or O? (x or o)").upper())
        
        if mark1 == 'X':
            player1 = 'X'
            player2 = 'O'
        else:
            player2 = 'X'
            player1 = 'O'
        
    return player1, player2 

    #to be reassigned as player markers later when bringing code together 







```

```python
#Ask for & validate position entry 

def ask_pos (person, marker):
    
    position = 0

    while position not in range(1,10):
        
        try:
            
            position = int(input(f"{person} Where will you place your {marker}? (1-9)"))

            if position>9 or position<1:
                print("Between 1 to 9 only. Try again.")
            else:
                break 
                
        except ValueError:
            print("Integers only.")
            continue 
            
    return position 
```

```python
#Append position entry to the table 


def table_edit (table,position,mark):
    
    table[position] = mark 
    return display(table)
```

```python
#Check for win -> return True if win // not True becomes tie 

def win_check(table, mark):
    
    if ((table[1]==mark and table[2]==mark and table[3]==mark) or 
    (table[4] == mark and table[5] == mark and table[6] == mark) or
    (table[7] == mark and table[8] == mark and table[9] == mark) or
    (table[1] == mark and table[5] == mark and table[9] == mark) or
    (table[7] == mark and table[5] == mark and table[3] == mark) or
    (table[1] == mark and table[4] == mark and table[7] == mark) or
    (table[2] == mark and table[5] == mark and table[8] == mark) or
    (table[3] == mark and table[6] == mark and table[9] == mark)): 
        
        return True
    
    else:
        
        return False
    
    







```

```python
#Option to restart game - return True or False

def restart ():
    
    prog_restart = ''
    progcheck = False
    
    while progcheck == False:
    
        if prog_restart!='y' or prog_restart!='n':
            prog_restart = input("Restart the game? (y or n)")
        
        if prog_restart=='y':
            progcheck = True
            return True 
        elif prog_restart=='n':
            progcheck = True
            return False
        else:
            continue
        
```

```python
def tie_check(table,mark):
        
        if ((table[1]=='X' or table[1]=='O') and (table[2]=='X' or table[2]=='O') and
        (table[3]=='X' or table[3]=='O') and (table[4]=='X' or table[4]=='O') and
        (table[5]=='X' or table[5]=='O') and (table[6]=='X' or table[6]=='O') and
        (table[7]=='X' or table[7]=='O') and (table[8]=='X' or table[8]=='O') and
        (table[9]=='X' or table[9]=='O')) and (not win_check(table,mark)):
            return True
        
        else:
            return False
    
```

```python
def final_game():

    gameon = True

    #showtable
    sampletable = ['']*10
    display(sampletable)

    #assign marker 
    player1, player2 = ass_marker()


    #ready to play?
    userin = input("Are you ready to play tic tac toe?")
    if userin == 'y':

        P1 = True

        while gameon == True:

            if P1==True: 

                    marker = player1
                    person1 = "P1: "

                    #ask input 
                    position = ask_pos(person1, player1)

                    #place input using player1 
                    table_edit (sampletable,position,marker)

                    #wincheck #continue to other player 
                    if win_check(sampletable, marker):
                        gameon = False
                        print ("Congratulations P1!")
                        break

                    elif tie_check(sampletable, marker):
                        print ("It's a tie.")
                        break

                    else:
                        P1=False

            if P1==False:

                    marker = player2
                    person2 = "P2: "

                    #ask input 
                    position = ask_pos(person2, player2)

                    #place input using player1 
                    table_edit (sampletable,position,marker)

                    #wincheck #continue to other player 
                    if win_check(sampletable, marker):
                        gameon = False
                        print ("Congratulations P2!")
                        break

                    elif tie_check(sampletable, marker):
                        print ("It's a tie.")
                        break
                        
                    else:
                        P1=True

        print ("Match over.")

    else:
        gameon = False
        
final_game()
```
